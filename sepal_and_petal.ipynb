```python
import pandas as pd
import seaborn as sns

iris = sns.load_dataset('iris')

sepal_petal_data = iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]

display(sepal_petal_data.head())
```

```python
import matplotlib.pyplot as plt

x = sepal_petal_data['sepal_length']
y = sepal_petal_data['sepal_width']

plt.scatter(x, y)
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.title('Scatter plot of sepal length vs sepal width')
plt.show()
```
